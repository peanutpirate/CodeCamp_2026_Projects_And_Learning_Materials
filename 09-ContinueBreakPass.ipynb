```python
benimListem = [5,10,15,20,25,30]
```

```python
for numara in benimListem:
    print(numara / 5)
```

```python
for numara in benimListem:
    if numara == 15:
        break
    print(numara)
```

```python
# belli bir koşulda döngüyü bitirmek istiyorsak breaki kullanıyoruz. 
```

```python
for numara in benimListem:
    if numara == 15:
        continue
    print(numara)
```

```python
for numara in benimListem:
    pass # ya orada bir işlem yapmıycaksak ya da sonradan yazıcaksak. kod yazarken karar verememişsen. 
```
