```python
## 2. Book SÄ±nÄ±fÄ±
# KitaplarÄ± temsil eden sÄ±nÄ±f
class Book:
    def __init__(self, title, author, year, is_available=True):
        self.title = title
        self.author = author
        self.year = year
        self.is_available = is_available

    def get_info(self):
        print("Kitap Bilgisi")
        print(f"BaÅŸlÄ±k: {self.title}")
        print(f"Yazar: {self.author}")
        print(f"YayÄ±n YÄ±lÄ±: {self.year}")
        print(f"Durum: {'Mevcut' if self.is_available else 'Ã–dÃ¼nÃ§te'}")

# Ã–rnek kullanÄ±m
book1 = Book("1984", "George Orwell", 1949)
book1.get_info()
```

```python
## 4.Default DeÄŸer KullanÄ±mÄ±
# Kitap sÄ±nÄ±fÄ± - is_available parametresinin default deÄŸer kullanÄ±mÄ±nÄ± gÃ¶sterir.
class Book:
    # is_available=True diyerek bir varsayÄ±lan deÄŸer atÄ±yoruz.
    def __init__(self, title, author, is_available=True):
        self.title = title
        self.author = author
        self.is_available = is_available

    def display_info(self):
        status = "MÃ¼sait" if self.is_available else "Ã–dÃ¼nÃ§ Verildi"
        print(f"Kitap: {self.title} | Durum: {status}")

# Uygulama Ã–rnekleri

# 1. Senaryo: Default DeÄŸeri Kullanma
# Yeni bir kitap aldÄ±k, rafta duruyor. is_available gÃ¶ndermemize gerek yok.
kitap1 = Book("Nutuk", "Mustafa Kemal AtatÃ¼rk")
kitap1.display_info()
# Ã‡Ä±ktÄ±: Kitap: Nutuk | Durum: MÃ¼sait

# 2. Senaryo: Default DeÄŸeri Ezme (Override)
# EÄŸer sisteme zaten birinde olan (kayÄ±t dÄ±ÅŸÄ± Ã¶dÃ¼nÃ§ verilmiÅŸ) bir kitap giriyorsak:
kitap2 = Book("SuÃ§ ve Ceza", "Dostoyevski", False)
kitap2.display_info()
# Ã‡Ä±ktÄ±: Kitap: SuÃ§ ve Ceza | Durum: Ã–dÃ¼nÃ§ Verildi

# 3. Neden Default KullanÄ±yoruz?
def create_library():
    print("\n--- Toplu Kitap GiriÅŸi ---")
    # Her seferinde True yazmak zahmetli ve hata payÄ±nÄ± artÄ±rÄ±r.
    kitaplar = [
        Book("Sefiller", "Victor Hugo"),
        Book("1984", "George Orwell"),
        Book("SimyacÄ±", "Paulo Coelho")
    ]

    for b in kitaplar:
        b.display_info()

create_library()
```

```python
## 5. Inheritance (DigitalBook)
# Parent (Ãœst) SÄ±nÄ±f: Genel kitap Ã¶zelliklerini barÄ±ndÄ±rÄ±r.
class Book:

    def __init__(self, title, author, is_available=True):
        self.title = title
        self.author = author
        self.is_available = is_available

    def display_info(self):
        status = "MÃ¼sait" if self.is_available else "Ã–dÃ¼nÃ§ Verildi"
        print(f"Kitap: {self.title} - Yazar: {self.author} - Durum: {status}")

# Child (Alt) SÄ±nÄ±f: Book sÄ±nÄ±fÄ±ndan miras alÄ±r.
# Dijital kitaplara Ã¶zel 'file_size' Ã¶zelliÄŸi ve 'download' metodu eklenmiÅŸtir.
class EBook(Book):

    def __init__(self, title, author, file_size, is_available=True):
        super().__init__(title, author, is_available)
        self.file_size = file_size  # MB cinsinden ek Ã¶zellik
    # E kitaplara Ã¶zel method
    def download(self):
        print(f"ğŸ“¥ '{self.title}' indiriliyor... (Boyut: {self.file_size} MB)")
    # Over write
    def display_info(self):
        status = "Ä°ndirmeye HazÄ±r" if self.is_available else "EriÅŸim KÄ±sÄ±tlÄ±"
        print(f"Dijital Kitap: {self.title} [{self.file_size}MB] - Durum: {status}")


# --- Uygulama ---

# 1. Standart Kitap OluÅŸturma (Parent)
fiziksel_kitap = Book("Ä°nce Memed", "YaÅŸar Kemal")
fiziksel_kitap.display_info()

# 2. E-Kitap OluÅŸturma (Child)
# Hem Book Ã¶zelliklerini (title, author) hem de kendi Ã¶zelliÄŸini (file_size) kullanÄ±r.
dijital_kitap = EBook("Veri Bilimi El KitabÄ±", "Jake VanderPlas", 15.5)
dijital_kitap.display_info()
dijital_kitap.download()
```
