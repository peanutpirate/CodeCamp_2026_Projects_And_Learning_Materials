```python
benimYemeklerim[1]
```

```python
# dictionary süslü parantezle çalışıyor anahtar kelime ve değeri yazmalıyız.
```

```python
benimSozluk = {"anahtarkelime" : "deger"}
```

```python
type(benimSozluk)
```

```python
benimSozluk["anahtarkelime"]
```

```python
benimYemekKaloriSozlugum = { "elma" : 100, "karpuz" : 200, "muz" : 300}
```

```python
benimYemekKaloriSozlugum["muz"]
```

```python
benimYemekKaloriSozlugum["elma"] = 200 # sözlügün içindeki değeri eşitliyerek değiştirebiliriz
```

```python
benimYemekKaloriSozlugum
```

```python
benimDegisikSozlugum = {150:"atıl",-2:"atlas"}
```

```python
benimDegisikSozlugum[150]
```

```python
yeniDictionary = {"anahtar1" : 100, "anahtar2" : [10,20,30,40,4.5,"atıl"], "anahtar3" : {"anahtar9" : 4}} 
```

```python
# sözlüğün içine sözlük de yazılabilir 
```

```python
yeniDictionary.keys()
```

```python
# .keys sözlüğün anahtarlarına ihtiyaç duyarsanız işe yarar
```

```python
# values sözlüğün değerlerini verir
```

```python
yeniDictionary.values()
```

```python
yeniDictionary["anahtar2"][-1]
```

```python
yeniDictionary["anahtar3"]["anahtar9"]
```
