## inheritance

```python
class Hayvan():
    def __init__(self):
        print("hayvan sınıfı init çağrıldı")

    def method1(self):
        print("hayvan sınıfı method1")

    def method2(self):
        print("hayvan sınıfı method2 çağrıldı")
```

```python
benimHayvanim = Hayvan()
```

```python
benimHayvanim.method1()
```

```python
class Kedi(Hayvan):
    def __init__(self):
        Hayvan.__init__(self)
        print("kedi sınıfı init çağrıldı")
```

```python
benimKedi = Kedi()
```

```python
benimKedi.method1()
```

```python
digerHayvan = Hayvan()
```

```python
digerKedi = Kedi()
```
