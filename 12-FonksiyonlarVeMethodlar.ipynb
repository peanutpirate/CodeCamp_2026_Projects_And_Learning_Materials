```python
def merhabaDunya(yazdirilacakIsim):
    print("merhaba")
    print(yazdirilacakIsim)
```

```python
merhabaDunya("atil")
```

## arg & kwargs

```python
# argümanlar ve kelime argümanları
```

```python
def yeniToplama(*args):
    return sum(args) # alınan bütün argümanları topla
    #arguments anlamında önündeki * önemli
```

```python
yeniToplama(10,20,30,40,50,60)
```

```python
def benimFonksiyonum(*args):
    return(args)
```

```python
benimFonksiyonum(20,30,40)
```

```python
# return edersek tuple olur. Return etmezsek nontype olur. 
```

```python
def ornekFonksiyon(**kwargs):
    return(kwargs)
```

```python
ornekFonksiyon(muz = 100, elma = 200, ananas = 300)
```

```python
def keyWordKontrolu(**kwargs):
    if "atil" in kwargs:
        print("atıl var")
    else:
        print("atıl yok")
```
