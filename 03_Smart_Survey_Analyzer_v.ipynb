## Smart Survey Analyzer (v3)

```python
# Rastgele Anket Verisi
import random

scores = []

while len(scores) < 20:
    score = random.randint(0, 6)

    if score == 0:
        break

    if score == 6:
        continue

    scores.append(score)
```

```python
# Veri Kontrolü
puanSayisi = len(scores)
print(f"Geçerli Puan Sayısı: ", (len(scores)))
print(f"Geçerli Puanlar: ", (scores))
# While koşul temellidir. While döngüsü koşul sağladığı sürece çalışır. Bu yüzden bu örnekte while döngüsü kullanılmıştır.
```

```python
# 3. Ortalama Hesaplama
ortalamaPuan = sum(scores) / puanSayisi
print(f"Ortalama Puan: {round(ortalamaPuan)}")
```

```python
# 4. Min / Max Analizi
min_puan = min(scores)
print(f"Minimum Puan: ", (min_puan))
max_puan = max(scores)
print(f"Maksimum Puan: ", (max_puan))
```

```python
# 5. Memnuniyet Analizi (For + If)
memnun = 0
kararsiz = 0
memnun_degil = 0

for score in scores:
    if score == 4 or score == 5:
        memnun += 1
    elif score == 3:
        kararsiz += 1
    elif score == 1 or score == 2:
        memnun_degil += 1
    else:
        pass

print("Memnun:", memnun)
print("Kararsız:", kararsiz)
print("Memnun değil:", memnun_degil)
```

```python
# 6. Erken Uyarı Sistemi (Break)
sayac=0
for score in scores:
    if score==1:
        sayac=sayac+1
    else:
        continue
    if sayac==3:
        print("⚠️ Kritik memnuniyetsizlik tespit edildi")
        break
```

```python
# 7. Genel Özet
#Genel Özet
print("Smart Survey Analyzer (v3)")
print()
print("Toplam geçerli cevap:", puanSayisi)
print("Ortalama puan:", round(ortalamaPuan))
print("Min puan:", min_puan)
print("Max puan:", max_puan)
print()
print("Memnun:", memnun)
print("Kararsız:", kararsiz)
print("Memnun değil:", memnun_degil)
```
