# zip

```python
yemekListesi = ["muz", "ananas", "elma"]
kaloriListesi = [100,200,300]
gunListesi = ["pazartesi","salı","çarşamba"]
```

```python
ziplenmisListe = list(zip(yemekListesi,kaloriListesi,gunListesi))
```

```python
for eleman in ziplenmisListe:
    print(type(eleman))
```

```python
ziplenmisListe
```
