## Local,Enclosing,Global,Built-In

```python
# LEGB model deniyor.
```

```python
benimAdim = "Ipek"
# global değişken
```

```python
def benimFonksiyonum():
    benimAdim = "mahmut"
    #enclosing
    def icFonksiyon():
        benimAdim = "ayşe"
        #local
        print(benimAdim)
    icFonksiyon()
```

```python
benimFonksiyonum()
```

```python
y = 10
def yeniFonksiyon(y):
    print(y)
    y = 5
    print(y)
    return y
```

```python
yeniFonksiyon(3)
```

```python
y
```

```python
y = yeniFonksiyon(3)
```

```python
y = 10

def ornekFonksiyon():
    global y
    y = 5
    print(y)
```

```python
ornekFonksiyon()
```
