## Smart Order Management System

```python
# 1. Ham Sipariş Verisi
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

```python
# 2. Sipariş Sayısı Analizi
total_orders = 0
paid = 0
unpaid = 0
for order in orders:
    total_orders += 1

    if order["is_paid"] == True:
        paid += 1
    else:
        unpaid += 1

print(f"Toplam sipariş sayısı: {len(orders)}")
print(f"Ödeme yapılmış sipariş sayısı: {paid}")
print(f"Ödeme yapılmamış sipariş sayısı: {unpaid}")
```

```python
# 3. Müşteri Listesi
all_customers = []
for order in orders:
    all_customers.append(order["customer"])

unique_customers_set = set(all_customers)
clean_customer_list = list(unique_customers_set)
print(f"Müşteri Listesi: ",clean_customer_list)
```

```python
# 4. Ürün Analizi
all_products = []
for order in orders:
    all_products.extend(order["items"])

print(f"Tüm Ürünler (Tekrarlı Liste): ", all_products)

unique_products_set = set(all_products)
clean_product_list = list(unique_products_set)
print(f"Tekrarsiz Ürünler: ", clean_product_list)
```

```python
# 5. Sipariş Durumu Kontrolü
if orders[0]["is_paid"]:
    print("ilk sipariş onaylandı")
else:
    print("ilk sipariş için ödeme bekleniyor")

if orders[1]["is_paid"]:
    print("ikinci sipariş onaylandı")
else:
    print("ikinci sipariş için ödeme bekleniyor")

if orders[2]["is_paid"]:
    print("üçüncü sipariş onaylandı")
else:
    print("üçüncü sipariş için ödeme bekleniyor")

if orders[3]["is_paid"]:
    print("dördüncü sipariş onaylandı")
else:
    print("dördüncü sipariş için ödeme bekleniyor")
```

```python
# 6. Sipariş Özeti
siparis1 = (orders[0]["customer"], orders[0]["total"])
siparis2 = (orders[1]["customer"], orders[1]["total"])
siparis3 = (orders[2]["customer"], orders[2]["total"])
siparis4 = (orders[3]["customer"], orders[3]["total"])
print("sipariş 1:", siparis1)
print("sipariş 2:", siparis2)
print("sipariş 3:", siparis3)
print("sipariş 4:", siparis4)
```

```python
# Yorum: Farklı birisi bu veriyi kullanırken müşteri adlarına ve ödediği fiyatlara müdahale edemez. Tuple'lar Immutable dır.
```

```python
# 7. Genel Çıktılar
print("Toplam sipariş:", len(orders))
print("Ödenmiş sipariş:", paid)
print("Ödenmemiş sipariş:",unpaid,"\n")
print("Unique ürünler:",clean_product_list)
print("Sipariş özetleri:",siparis1, siparis2, siparis3, siparis4)
```
