```python
# Döngü: Bir işlemi devamlı yapmamımıza yarayan sayıdır. 
```

```python
benimListem = [10,20,30,40,50]
```

```python
if 10 in benimListem:
    print("evet içerisinde mevcut")
```

```python
benimListem[0] * 5 / 3
```

```python
benimListem[1] * 5 / 3
```

```python
print("döngü başladı")
for numara in benimListem:
    print(numara * 5 / 3)
print("döngü bitti")
```

```python
# for ... bir değişken ismi ve for dongusune soktugum için benim listemdeki tüm elemanları numara değişkenine atıyor. 
# döngü dememizin bir sebebi de bu bir tab boşluk döngü içinde yapılan şeyleri anlatıyor. 
```

```python
for num in benimListem:
    yeniNumara = num * 5 / 3
    print(yeniNumara)
```

```python
yeniListe = [1,2,3,4,5,6]
```

```python
for rakam in yeniListe:
    if rakam % 2 == 0: #kalanını bulmak ikiye bölünebiliyorsa kalanını bulmak ikiye böldüğümüzde sıfır kalsın.
        print(rakam) # for loopun içinde istediğimiz kontrolu yapabiliyoruz.
```

```python
yeniString = "Ipek Egmen"
```

```python
for harf in yeniString:
    print(harf)
```
