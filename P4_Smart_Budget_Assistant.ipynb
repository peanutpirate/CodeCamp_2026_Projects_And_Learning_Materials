# ğŸ§© KULLANILAN PYTHON KONULARI
Bu projede yalnÄ±zca aÅŸaÄŸÄ±daki Python konularÄ± kullanÄ±lmaktadÄ±r:
- Fonksiyonlar
- Girdi / Ã‡Ä±ktÄ± iÅŸlemleri (input, print)

- Ã–nemli gÃ¶mÃ¼lÃ¼ fonksiyonlar:

```python
# 2. Harcama Alma Fonksiyonu
def get_expenses():
    # KullanÄ±cÄ±dan gÃ¼venli ÅŸekilde harcama listesi alÄ±r
    expenses = []

    # Harcama sayÄ±sÄ±nÄ± al
    while True:
        try:
            n = int(input("KaÃ§ harcama gireceksiniz? "))
            if n <= 0:
                print("LÃ¼tfen 1 veya daha bÃ¼yÃ¼k bir sayÄ± girin.")
            else:
                break
        except ValueError:
            print("GeÃ§ersiz giriÅŸ. LÃ¼tfen tam sayÄ± girin.")

    i = 0
    # HarcamalarÄ± tek dÃ¶ngÃ¼yle al
    while i < n:
        try:
            val = float(input(f"{i+1}. harcama tutarÄ±: "))
            if val < 0:
                print("Harcama tutarÄ± negatif olamaz.")
            else:
                expenses.append(val)
                i += 1
        except ValueError:
            print("GeÃ§ersiz giriÅŸ. LÃ¼tfen sayÄ±sal bir deÄŸer girin.")

    return expenses

harcamalar = get_expenses()
print("Girilen harcamalar:", harcamalar)
```

```python
# 3. Toplam ve Ortalama Fonksiyou
def calculate_summary(expenses):
    # Toplam ve ortalama harcamayÄ± hesaplar

    if len(expenses) == 0:
        print("Harcama listesi boÅŸ.")
        return

    total = sum(expenses)
    average = total / len(expenses)

    print("Toplam Harcama:", round(total, 2))
    print("Ortalama Harcama:", round(average, 2))

harcamalar = get_expenses()
calculate_summary(harcamalar)
```

```python
# 4.*Args ile Esnek Hesaplama
def flexible_total(*args):
    # Girilen tÃ¼m argÃ¼manlarÄ± toplar ve argÃ¼man sayÄ±sÄ±nÄ± gÃ¶sterir

    if len(args) == 0:
        print("HiÃ§ harcama girilmedi.")
        return

    total = sum(args)

    print("Girilen harcama sayÄ±sÄ±:", len(args))
    print("Toplam Harcama:", round(total, 2))
  # Ã–rnek
flexible_total(120, 300, 75.5)
```

```python
# 5. En BÃ¼yÃ¼k/En KÃ¼Ã§Ã¼k Harcama
def find_extremes(expenses):
    # En bÃ¼yÃ¼k ve en kÃ¼Ã§Ã¼k harcamayÄ± bulur

    if len(expenses) == 0:
        print("Harcama listesi boÅŸ.")
        return

    max_expense = max(expenses)
    min_expense = min(expenses)

    print("En BÃ¼yÃ¼k Harcama:", round(max_expense, 2))
    print("En KÃ¼Ã§Ã¼k Harcama:", round(min_expense, 2))

harcamalar = get_expenses()
find_extremes(harcamalar)
```

```python
# 6. Kapsam (Scope) GÃ¶sterimi - Ã–nemli!

# currency deÄŸiÅŸkeni global olarak tanÄ±mlanmÄ±ÅŸtÄ±r
# Fonksiyon iÃ§inde bu deÄŸiÅŸken okunabilir

# Ancak fonksiyon iÃ§inde currency = "USD" gibi bir atama yapÄ±lÄ±rsa
# Python bunu global deÄŸil, local deÄŸiÅŸken olarak algÄ±lar

# Bu nedenle:
# - Global deÄŸiÅŸken okunabilir
# - Global deÄŸiÅŸken doÄŸrudan deÄŸiÅŸtirilemez
# - AynÄ± isimle yapÄ±lan atama local scope oluÅŸturur

# AmaÃ§:
# Local vs Global scope farkÄ±nÄ± gÃ¶stermek

currency = "TL"

def show_currency():
    # Global deÄŸiÅŸkeni OKUMA
    print("Fonksiyon iÃ§inden okunan para birimi:", currency)

    # currency = "USD"
    # print("Fonksiyon iÃ§inde deÄŸiÅŸtirilen para birimi:", currency)

show_currency()

print("Fonksiyon Ã§aÄŸrÄ±sÄ±ndan sonra global currency:", currency)
```

```python
# 7. Ana AkÄ±ÅŸ (Main Flow)
print("=== Smart Budget Assistant (v4) ===")

# 1ï¸âƒ£ HarcamalarÄ± al
harcamalar = get_expenses()

# 2ï¸âƒ£ Toplam ve ortalama
print("\n--- Harcama Ã–zeti ---")
calculate_summary(harcamalar)

# 3ï¸âƒ£ En bÃ¼yÃ¼k / en kÃ¼Ã§Ã¼k harcama
print("\n--- Harcama Limitleri ---")
find_extremes(harcamalar)

# 4ï¸âƒ£ Esnek toplam (args gÃ¶sterimi)
print("\n--- Esnek Toplam (Args Demo) ---")
flexible_total(*harcamalar)

print("\nProgram tamamlandÄ± âœ…")
```
